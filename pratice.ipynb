```python
import torch.nn as nn
class linearreg(nn.Module):
    def __init__(self):
        super().__init__()
        self.weight = nn.Parameter(torch.randn(1,requires_grad = True , dtype = torch.float))
        self.bias = nn.Parameter(torch.randn(1,requires_grad = True , dtype = torch.float))
    def forward(self , x : torch.Tensor) -> torch.Tensor:
        return (x * self.weight) + self.bias
```

```python
class blobmodel(nn.Module):
    def __init__(self , input_features , output_features , hidden_layer):
        super().__init__()
        self.stack = nn.Sequential(
                        nn.Linear(in_features = input_features , out_features = hidden_layer ),
                        nn.Linear(in_features = hidden_layer , out_features = hidden_layer),
                        nn.Linear(in_features = hidden_layer , out_features = output_features)
                        
        )
    def forward(self , x: torch.Tensor)-> torch.Tensor:
        return self.stack(x)
```

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
torch.manual_seed(19)
NUM_FEATURES = 2
NUM_CLASS = 4

x_blob, y_blob = make_blobs(
    n_samples=1000,
    n_features=NUM_FEATURES,
    centers=NUM_CLASS,
    cluster_std=1.5,
    random_state=19
)


x_blob = torch.from_numpy(x_blob).type(torch.float32)
y_blob = torch.from_numpy(y_blob).type(torch.long)

x_blob_train, x_blob_test, y_blob_train, y_blob_test = train_test_split(
    x_blob, y_blob, test_size=0.2, random_state=19
)

x_blob_train = x_blob_train.to(device)
x_blob_test = x_blob_test.to(device)
y_blob_train = y_blob_train.to(device)
y_blob_test = y_blob_test.to(device)

class BlobModel(nn.Module):
    def __init__(self, input_features, hidden_units, output_features):
        super().__init__()
        self.stack = nn.Sequential(
            nn.Linear(input_features, hidden_units),
            nn.ReLU(),
            nn.Linear(hidden_units, hidden_units),
            nn.ReLU(),
            nn.Linear(hidden_units, output_features)
        )

    def forward(self, x):
        return self.stack(x)

model = BlobModel(NUM_FEATURES, 16, NUM_CLASS).to(device)

loss_fn = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)

acc_fn = Accuracy(task="multiclass", num_classes=NUM_CLASS).to(device)

epochs = 200

for epoch in range(epochs):
    
    model.train()
    
    y_logits = model(x_blob_train)
    y_pred = torch.argmax(y_logits, dim=1)
    
    loss = loss_fn(y_logits, y_blob_train)
    acc = acc_fn(y_pred, y_blob_train)
    
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    
    model.eval()
    with torch.inference_mode():
        y_logits_test = model(x_blob_test)
        y_pred_test = torch.argmax(y_logits_test, dim=1)
        
        loss_test = loss_fn(y_logits_test, y_blob_test)
        acc_test = acc_fn(y_pred_test, y_blob_test)

    if epoch % 20 == 0:
        print(f"Epoch {epoch} | Train Loss: {loss:.4f} | Test Loss: {loss_test:.4f} | Train Acc: {acc:.4f} | Test Acc: {acc_test:.4f}")
```

```python
class MNISTv0(nn.Module):
    def __init__(self , input_shape : int , hidden_units : int , output_shape : int):
        super().__init__()
        self.layer_stack = nn.Sequential( nn.Flatten(),
                                          nn.Linear(in_features = input_shape , out_features = hidden_units),
                                          nn.Linear(in_features = hidden_units , out_features = output_shape),
                                          )
    def forward(self , x : torch.Tensor ) -> torch.Tensor:
        return self.layer_stack(x)
```

```python
m0 = MNISTv0(input_shape = 784 , hidden_units = 10 , output_shape = len(classes)).to(device)
```

```python
from torchmetrics import Accuracy 
loss_fn = nn.CrossEntropyLoss()
optimizer = torch.optim.SGD(params = m0.parameters() , lr = 0.01)
```

```python
from tqdm.auto import tqdm
torch.manual_seed(19)

device = "cuda" if torch.cuda.is_available() else "cpu"
m0 = m0.to(device)

epochs = 10

for epoch in range(epochs):
    print(f"\nEpoch: {epoch+1}/{epochs}")
    
    
    train_loss = 0
    
    for batch, (x, y) in tqdm(enumerate(train_dataloader), total=len(train_dataloader)):
        x, y = x.to(device), y.to(device)
        
        m0.train()
        y_pred = m0(x)
        loss = loss_fn(y_pred, y)
        
        train_loss += loss.item()
        
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
    
    train_loss /= len(train_dataloader)

    
    test_loss = 0
    m0.eval()
    
    with torch.inference_mode():
        for x, y in test_dataloader:
            x, y = x.to(device), y.to(device)
            
            test_pred = m0(x)
            loss_test = loss_fn(test_pred, y)
            
            test_loss += loss_test.item()
    
    test_loss /= len(test_dataloader)

    
    print(f"Train loss: {train_loss:.4f} | Test loss: {test_loss:.4f}")
```

```python
def evaluate(model: nn.Module, dataloader: torch.utils.data.DataLoader, loss_fn: nn.Module):
    
    model.eval()
    loss = 0
    
    with torch.inference_mode():
        for x, y in dataloader:
            x, y = x.to(device), y.to(device) 
            
            test_pred = model(x)
            batch_loss = loss_fn(test_pred, y)
            
            loss += batch_loss.item()  
    
    loss /= len(dataloader) 
    
    return {
        "model": model.__class__.__name__,
        "loss": loss
    }
```

```python
class MNISTv3(nn.Module):
    def __init__(self , input_feature : int , hidden_feature : int , output_feature : int):
        super().__init__()
        self.block1 = nn.Sequential(nn.Conv2d(in_channels = input_feature , out_channels = hidden_feature , kernel_size = 3 , stride = 1 , padding = 1),
                                    nn.ReLU(),
                                    nn.Conv2d(in_channels = hidden_feature , out_channels = hidden_feature , kernel_size = 3 , stride = 1 , padding = 1),
                                    nn.ReLU(),
                                    nn.MaxPool2d(kernel_size = 2 , stride = 2))
        self.block2 = nn.Sequential(nn.Conv2d(in_channels = hidden_feature , out_channels = hidden_feature , kernel_size = 3 , stride = 1 , padding = 1),
                                    nn.ReLU(),
                                    nn.Conv2d(in_channels = hidden_feature , out_channels = hidden_feature , kernel_size = 3 , stride = 1 , padding = 1),
                                    nn.ReLU(),
                                    nn.MaxPool2d(kernel_size = 2 , stride = 2))
        self.block3 = nn.Sequential(nn.AdaptiveAvgPool2d((1,1)),
                                    nn.Flatten(),
                                    nn.Linear(in_features = hidden_feature , out_features = output_feature))
    def forward(self , x):
        return self.block3(self.block2(self.block1(x)))
```

```python
def train_step(model: torch.nn.Module, dataloader: torch.utils.data.DataLoader, loss_fn: torch.nn.Module, optimizer: torch.optim.Optimizer):
    model.train()
    train_loss = 0
    for batch, (x, y) in enumerate(dataloader):
        x, y = x.to(device), y.to(device)
        y_pred = model(x)
        loss = loss_fn(y_pred, y)
        train_loss += loss.item() 
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
    return train_loss / len(dataloader)
```

```python
def test_step(model: torch.nn.Module, dataloader: torch.utils.data.DataLoader, loss_fn: torch.nn.Module):
    model.eval() 
    test_loss =  0
    with torch.inference_mode():
        for batch, (x, y) in enumerate(dataloader):
            x, y = x.to(device), y.to(device)
            test_pred_logits = model(x)
            loss = loss_fn(test_pred_logits, y)
            test_loss += loss.item()
    return test_loss / len(dataloader)
```

```python
from tqdm.auto import tqdm
def train(model: torch.nn.Module, train_dataloader: torch.utils.data.DataLoader,  test_dataloader: torch.utils.data.DataLoader,  optimizer: torch.optim.Optimizer, loss_fn: torch.nn.Module, epochs: int = 5):
    results = {"train_loss": [], "test_loss": []}
    for epoch in tqdm(range(epochs)):
        model.train()
        train_loss = 0
        for batch, (x, y) in enumerate(train_dataloader):
            x, y = x.to(device), y.to(device)
            y_pred = model(x)
            loss = loss_fn(y_pred, y)
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
            train_loss += loss.item()

        test_loss = 0
        model.eval()
        with torch.inference_mode():
            for batch, (x, y) in enumerate(test_dataloader):
                x, y = x.to(device), y.to(device)
                test_pred = model(x)
                loss = loss_fn(test_pred, y)
                test_loss += loss.item()
        train_loss = train_loss / len(train_dataloader)
        test_loss = test_loss / len(test_dataloader)
        print(f"Epoch =  {epoch}  train_loss =  {train_loss}  test_loss = {test_loss}")
        results["train_loss"].append(train_loss)
        results["test_loss"].append(test_loss)
    return results
```
